```python
import urllib
import httplib2

qparams = {"min_date":"201510","max_date":"201511"}
parqparams = urllib.parse.urlencode(qparams)

conn = httplib2.HTTPSConnectionWithTimeout('apis.bbva.com')

headers = {}
headers["Authorization"] = "jwt ******" # change for your token when download

conn.request('GET', '/paystats_download_sbx/1/datasets?%s'%parqparams, None, headers)

r = conn.getresponse()
data = r.read()
print(r.status, r.read())
```
